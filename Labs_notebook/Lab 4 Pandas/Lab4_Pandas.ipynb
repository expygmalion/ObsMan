```python
Q1 = df['Doors'].quantile(0.25)
Q3 = df['Doors'].quantile(0.75)

IQR = Q3 - Q1

print(Q1, Q3, IQR)
```

```python
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

print(lower_bound, upper_bound)

outliers = df[(df['Doors'] < lower_bound) | (df['Doors'] > upper_bound)]

new_df = df[(df['Doors'] >= lower_bound) & (df['Doors'] <= upper_bound)]

new_df



```
