# KNN

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple synthetic dataset for classification
X_train, y_train = make_classification(n_samples=100, n_features=20,
                                       n_classes=2, random_state=42)

# Test K values from 1 to 20
k_values = range(1, 21)
mean_accuracies = []
```

```python
# Perform k-fold cross-validation for each K value
for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    # 5-fold cross-validation
    accuracies = cross_val_score(model, X_train, y_train, cv=5)
    # Average accuracy across folds
    mean_accuracies.append(np.mean(accuracies))

# You can add plotting code here if you'd like to visualize the results
# plt.plot(k_values, mean_accuracies)
# plt.xlabel('Number of Neighbors (K)')
# plt.ylabel('Mean Cross-Validation Accuracy')
# plt.title('KNN Performance for Different K Values')
# plt.show()
```

```python
# Create a wider plot (Width = 10, Height = 4)
plt.figure(figsize=(10, 4))

# Plot the accuracies vs. K values
plt.plot(k_values, mean_accuracies, marker='o', color='b')
plt.xlabel('K Value')  # Label for the x-axis
plt.ylabel('Mean Accuracy')  # Label for the y-axis
plt.title('Accuracy vs. K Value')  # Title of the graph

# Set the x-ticks to display the K values clearly
plt.xticks(k_values)

# Annotate the mean accuracy values on the plot
for i, mean_accuracy in enumerate(mean_accuracies):
    plt.text(k_values[i], mean_accuracy,
             f'{mean_accuracy:.2f}', ha='center', va='bottom', fontsize=8)

# Show the plot
plt.grid(True)
plt.show()
```

# ID3

```python
#ID3
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Generate a simple synthetic dataset for classification
X, y = make_classification(n_samples=100, n_features=5,
                           n_classes=2, random_state=42)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=42)
```

```python
# Create a Decision Tree classifier 
#(ID3-like behavior, using entropy for information gain)
model = DecisionTreeClassifier(criterion='entropy', random_state=42)

# Train the model on the training data
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Plot the decision tree
plt.figure(figsize=(10, 12))
plot_tree(model, filled=True, feature_names=
          [f"Feature {i}" for i in range(X.shape[1])],
          class_names=['Class 0', 'Class 1'], rounded=True)
plt.title("Decision Tree (ID3-like) Visualization")
plt.show()
```
